```python
def convert_cat_var(cat_var):
    temp = data.groupby(cat_var)['target'].sum()/data.groupby(cat_var)['transaction_id'].count()
    temp=pd.DataFrame(temp)
    temp.columns=[['num'+cat_var]]
    temp[cat_var] = temp.index
    tempdata=data.merge(temp,how='left',on=cat_var)
    return tempdata
```

```python
def convert_cat_var2(cat_var):
    temp = data.groupby(cat_var)['target'].count()/data.shape[0]
    temp=pd.DataFrame(temp)
    temp.columns=[['num'+cat_var]]
    temp[cat_var] = temp.index
    tempdata=data.merge(temp,how='left',on=cat_var)
    return tempdata
```

```python
for l in range(1,19):
    data=convert_cat_var('cat_var_'+str(l))
    data=convert_cat_var2('cat_var_'+str(l))
```
